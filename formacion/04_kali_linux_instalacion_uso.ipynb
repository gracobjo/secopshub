## Simulación: IOC desde artefacto de lab

Cuando en Kali generes un hash (`sha256sum`), pégalo abajo para clasificar el tipo de indicador (misma idea que el notebook 02 / SecOps Hub).

````python
import re

def clasificar_ioc(valor: str) -> str:
    v = valor.strip().lower()
    if re.fullmatch(r"[0-9a-f]{64}", v):
        return "hash_sha256"
    if re.fullmatch(r"[0-9a-f]{32}", v):
        return "hash_md5"
    if re.fullmatch(r"(?:\d{1,3}\.){3}\d{1,3}", v):
        return "ipv4"
    if re.fullmatch(r"[a-z0-9.-]+\.[a-z]{2,}", v):
        return "dominio"
    return "desconocido"

ejemplos = [
    "e3b0c44298fc1c149afbf4c8996fb92427ae41e4649b934ca495991b7852b855",
    "127.0.0.1",
    "malicioso.ejemplo.lab",
]
for e in ejemplos:
    print(f"{e[:48]:48} → {clasificar_ioc(e)}")

mio = input("Pega aquí tu IOC (Enter para saltar): ").strip()
if mio:
    print("Tipo detectado:", clasificar_ioc(mio))
````
